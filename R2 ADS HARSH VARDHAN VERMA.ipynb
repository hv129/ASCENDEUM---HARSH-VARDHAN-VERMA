```python
df_agg = df.groupby("bidder").agg({"prebid_win_count": "sum", "request_count": "sum"}).reset_index()
```

```python
df_agg["win_rate"] = df_agg['prebid_win_count'] / df_agg['request_count']
```

```python
top2_win_rate = df_agg.sort_values("win_rate", ascending=False).head(2)
top2_win_rate
```
